```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB

# Definisanje putanje do fajlova
train_data_path = 'adult.data'
test_data_path = 'adult.test'

# Imena kolona prema opisu dataset-a
columns = ["age", "workclass", "fnlwgt", "education", "education-num", "marital-status", 
           "occupation", "relationship", "race", "sex", "capital-gain", "capital-loss", 
           "hours-per-week", "native-country", "income"]

# Učitavanje trening i test podataka
train_data = pd.read_csv(train_data_path, header=None, names=columns, skipinitialspace=True)
test_data = pd.read_csv(test_data_path, header=None, names=columns, skipinitialspace=True, skiprows=1)

train_data['income'] = train_data['income'].str.strip().str.replace('\.', '', regex=True)
test_data['income'] = test_data['income'].str.strip().str.replace('\.', '', regex=True)

# Zamena '?' sa NaN i čišćenje podataka od nedostajućih vrednosti
train_data.replace('?', pd.NA, inplace=True)
train_data.dropna(inplace=True)
test_data.replace('?', pd.NA, inplace=True)
test_data.dropna(inplace=True)

# Uklanjanje outlier-a u koloni 'capital-gain'
train_data = train_data[train_data['capital-gain'] != 99999]
test_data = test_data[test_data['capital-gain'] != 99999]

# Razdvajanje podataka na features i target
X_train = train_data.drop('income', axis=1)
y_train = train_data['income']
X_test = test_data.drop('income', axis=1)
y_test = test_data['income']

# Preprocesiranje numeričkih i kategorijalnih podataka
numeric_features = ['age', 'fnlwgt', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week']
categorical_features = ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'native-country']

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), categorical_features)
    ])

# Kreiranje pipeline-a za različite modele
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000, class_weight='balanced'),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced'),
    'KNN': KNeighborsClassifier(n_neighbors=5),
    'SVM': SVC(kernel='linear', C=1, probability=True),
    'Naive Bayes': GaussianNB()
}

for name, model in models.items():
    clf = Pipeline(steps=[('preprocessor', preprocessor),
                          ('classifier', model)])
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)
    report = classification_report(y_test, y_pred)

    print(f"{name} Model Accuracy: {accuracy}")
    print(f"{name} Classification Report:\n{report}")
    sns.heatmap(cm, annot=True, fmt="d", cmap='Blues', xticklabels=['<=50K', '>50K'], yticklabels=['<=50K', '>50K'])
    plt.title(f'{name} Confusion Matrix')
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.show()
```

```python
from sklearn.utils import resample

# Kombiniranje X_train i y_train u jedan DataFrame
train_data_resample = pd.concat([X_train, y_train], axis=1)

# Razdvajanje na dvije klase
majority_class = train_data_resample[train_data_resample.income == '<=50K']
minority_class = train_data_resample[train_data_resample.income == '>50K']

# Undersampling većinske klase
majority_class_downsampled = resample(majority_class, 
                                      replace=False,    # uzorkovanje bez zamjene
                                      n_samples=len(minority_class),     # broj uzoraka kao u manjoj klasi
                                      random_state=42)  # za reproduktivnost

# Kombinovanje undersample-ovane većinske klase sa manjinskom klasom
train_data_balanced = pd.concat([majority_class_downsampled, minority_class])

# Razdvajanje nazad na X i y
X_train_balanced = train_data_balanced.drop('income', axis=1)
y_train_balanced = train_data_balanced['income']

# Treniranje modela na balansiranim podacima
for name, model in models.items():
    clf = Pipeline(steps=[('preprocessor', preprocessor),
                          ('classifier', model)])
    clf.fit(X_train_balanced, y_train_balanced)
    y_pred = clf.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)
    report = classification_report(y_test, y_pred)
    
    print(f"{name} Model Accuracy: {accuracy}")
    print(f"{name} Classification Report:\n{report}")
    sns.heatmap(cm, annot=True, fmt="d", cmap='Blues', xticklabels=['<=50K', '>50K'], yticklabels=['<=50K', '>50K'])
    plt.title(f'{name} Confusion Matrix')
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.show()
```
